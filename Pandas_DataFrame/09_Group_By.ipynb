# Splitting

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("datasets/census.csv")

# Exclude state level summarization, wich have sum level value of 40
df = df[ df["SUMLEV"]==50 ]
df.head()
```

```python
len(df)
# Let's run such task for 3 times and time it. For this we'll use the cell magic function %%timeit
```

```python
%%timeit -n 3

for state in df["STNAME"].unique():
    # We'll just calculate the average using numpy
    avg = np.average(df.where(df["STNAME"]==state).dropna()["CENSUS2010POP"])
    # And we'll print it to the screen
    print("Counties in state:", state, "have an average population of", str(avg))
```

```python
# It takes a fair bit of time to finish, let's try another approach using gruopby()
```

```python
%%timeit -n 3

for group, frame in df.groupby("STNAME"):
    # groupby() returns a tuple, where the first value is the value of the key we were trying
    # to group by, in this case a specific name, and the second one is projected DataFrame 
    # that was found for that group
    
    # Now let's calculate the average of census2010pop
    avg = np.average(frame["CENSUS2010POP"])
    
    # Print the results
    print("Counties in state:", group, "have an average population of", str(avg))
    
```

Huge difference in speed. We could create some functuin which returns a number between zero an two, based on the first character of the state name. Then, we can tell gruopby() to use this function to split up our DataFrame. It's important to note that in order to do this, you need to set the index of the DataFrame to be the column that you want to group by first.

```python
df = df.set_index("STNAME")
```

```python
def set_batch_number(item):
    if item[0] < "M":
        return 0
    if item[0] < "Q":
        return 1
    return 2

for group, frame in df.groupby(set_batch_number):
    print("There are", str(len(frame)), "records in gruop", str(group), "for processing.")
```

```python
# Notice that this time I didn't pass in a column name to groupby(), instead, I set the index of the DataFrame 
# to be STNAME, and if no column identifier is passed gruopby() will automatically use the index.
```
