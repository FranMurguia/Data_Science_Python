```python
# Data cleaning
import pandas as pd

df = pd.read_csv("Datasets/US_Presidents_1.csv")

df.head()
```

```python
df.columns
```

```python
# Split the name into Firstname and Lastname

# Copy the column with the names
df["First"] = df["President "]
# Call replace()
df["First"] = df["First"].replace("[ ].*", "", regex = True)

df.head()
```

```python
# Works but its slow, so, I want to use the apply() function.

# Define a function which splits a string into two pieces using a single row of data and a key name
def split_name(row):
    # Extract the first name
    row["First"] = row["President "].split(" ")[0]
    # Extract the last name
    row["Last"] = row["President "].split(" ")[-1]
    return row

df = df.apply(split_name, axis = "columns")
df.head()
```

```python
# Lets use the extract function
del(df["First"])
del(df["Last"])
```

```python
# Extract takes a regular expression as input and specifically requires you to set capture groups that 
# correspond to the output columns you are intesested in.

# Match 3 groups but only return 2:
# 1st: Any characters or digits, any number of those
# 2nd: A question mark colon because I want this group not to be returned and I want any number of 
# characters followed by a white space
# 3rd Any number of characters, but I want to anchor this to the back of the string
pattern = "(^[\w]*)(?:.* )([\w]*$)"

df["President "].str.extract(pattern).head()
```

```python
# Change the column names
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

names = df["President "].str.extract(pattern)

names.head()
```

```python
df["First"] = names["First"]
df["Last"] = names["Last"]

df.head()
```
