```python
record1 = pd.Series({"Name": "Francisco",
                    "Class": "Physics",
                    "Score": 85})
record2 = pd.Series({"Name": "Javier",
                    "Class": "Chemistry",
                    "Score": 82})
record3 = pd.Series({"Name": "Jonathan",
                    "Class": "English",
                    "Score": 90})

df = pd.DataFrame([record1, record2, record3],
                 index = ["school 1", "school 2", "school 3"])
df
```

```python
df.rename(mapper=lambda x: x.upper(), axis=1, inplace=True)
df
```

```python
df[( df["toefl score"].isin(range(106,115)) )]
```

```python
df[ df["toefl score"].gt(105) & df["toefl score"].lt(115) ]
```
