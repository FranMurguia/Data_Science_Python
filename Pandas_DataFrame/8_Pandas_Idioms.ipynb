```python
import pandas as pd
import numpy as np
import timeit
```

```python
df = pd.read_csv("datasets/Census.csv")
df.head()
```

## Pandorable

```python
# First we use the where() function and pass in a boolean mask wich is only true for those rows where
# the SUMLEV is equal to 50.
# With the result of the where() function evaluated, we drop missing values.
# Next, we set an index on the result of that. In this case, I've set it to state name followed by
# the county name.
# Finally, rename a column to make it more readable.
# Note: I began the statement with a parenthesis, which tells python I'm going to span the statement
# over multiple lines, for readability.

(df.where( df["SUMLEV"]==50 )
 .dropna()
 .set_index(["STNAME", "CTYNAME"])
 .rename(columns={"ESTIMATESBASE2010": "Estimates Base 2010"}))
```

## Non-Pandorable

```python
# First create a new dataframe from the original
df = df[ df["SUMLEV"]==50 ] # I'll use the overloaded indexing operator [] wich frops nan's
# Update index
df.set_index(["STNAME", "CTYNAME"], inplace=True)
# Set the colum names
df.rename(columns={"ESTIMATESBASE2010": "Estimates Base 2010"})
```

## Execution time

```python
def first_approach():
    global df
    return (df.where(df["SUMLEV"]==50)
            .dropna()
            .set_index(["STNAME", "CTYNAME"])
            .rename(columns={"ESTIMATESBASE2010": "Estimates Base 2010"}))

df = pd.read_csv("Datasets/Census.csv")

timeit.timeit(first_approach, number=10)
```

```python
def second_approach():
    global df
    new_df=df[ df["SUMLEV"]==50 ]
    new_df.set_index(["STNAME", "CTYNAME"])
    return new_df.rename(columns={"ESTIMATESBASE2010": "Estimates Base 2010"})

df = pd.read_csv("Datasets/Census.csv")

timeit.timeit(second_approach, number=10)
```

The second approach is much faster! So, this is a particular example of a classic time readability trade off.

There is a lots of examples on stack overflow and in documentation of people using method chaining in their pandas. And so, being able to read and understand the syntax is really worth your time. But it's important to keep in mind that following that appears to be stylistic idioms might have performance issues that you need to consider as well.

## Apply Map

```python
# In census DataFrame we have five columns for population estimates, with each corresponding 
# with one year of estimates. Let's create some new columns for minimum or maximum values, 
# and the apply functions is an easy way to do this.

# First write the function
def min_max(row):
    data = row[["POPESTIMATE2010",
              "POPESTIMATE2011",
              "POPESTIMATE2012",
              "POPESTIMATE2013",
              "POPESTIMATE2014",
              "POPESTIMATE2015"]]
    
    return pd.Series({"Min": np.min(data), "Max": np.max(data)})
```

```python
df = pd.read_csv("Datasets/Census.csv")
df.columns
```

```python
# Then we just need to call apply on the DataFrame

df.apply(min_max, axis="columns").head()
```

```python
# Now we add two new columns in the original DataFrame to store Min and Max

def min_max(row):
    # Extract data
    data = row[ ["POPESTIMATE2010",
                "POPESTIMATE2011",
                "POPESTIMATE2012",
                "POPESTIMATE2013",
                "POPESTIMATE2014",
                "POPESTIMATE2015"] ]
    
    # Create a new entry for max
    row["MAX"] = np.max(data)
    
    # Create a new entry for min
    row["MIN"] = np.min(data)
    
    return row
```

```python
# Now just apply the function across the DataFrame

df.apply(min_max, axis="columns").head()
```

```python
# Apply is rarely used eith large function definitions, like we did.
# Instead, is typically see it used with lambdas.

rows = ["POPESTIMATE2010",
        "POPESTIMATE2011",
        "POPESTIMATE2012",
        "POPESTIMATE2013",
        "POPESTIMATE2014",
        "POPESTIMATE2015"]

df.apply(lambda x: np.max(x[rows]), axis=1).head()
```

```python
def get_state_region(x):
    northeast = ["Connecticut", "Maine", "Massachusetts", "New Hampshire",
                 "Rhode Island", "Vermont", "New York", "New Jersey", "Pennsylvania"]
    
    midwest = ["Illinois", "Indiana", "Michigan", "Ohio", "Wisconsin", "Iowa", 
               "Kansas", "Minnesota", "Missouri", "Nebraska", "North Dakota", 
               "South Dakota"]
    
    south = ["Delaware", "Florida", "Georgia", "Maryland", "North Carolina", 
             "South Carolina", "Virginia", "District of Columbia", "West Virginia", 
             "Alabama", "Kentucky", "Mississippi", "Tennessee", "Arkansas", 
             "Louisiana" , "Oklahoma", "Texas"]
    
    west = ["Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico", "Utah", 
            "Wyoming", "Alaska", "California", "Hawaii", "Oregon", "Washington"]
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"
```

```python
df["state_region"] = df["STNAME"].apply(lambda x: get_state_region(x))
```
