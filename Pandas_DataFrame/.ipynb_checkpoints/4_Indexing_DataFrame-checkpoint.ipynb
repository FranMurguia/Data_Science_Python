```python
df = pd.read_csv("DataSets/Census.csv")
df.head()
```

```python
df["SUMLEV"].unique()
```

```python
# Exclude all of the rows that are summaries at the state level and just keep the county data.
df = df[df["SUMLEV"] == 50]
df.head()
```

```python
# Reduce the data at just the total population estimates and the total number of births.

df.columns
```

```python
columns_to_keep = ["STNAME", "CTYNAME", "BIRTHS2010", "BIRTHS2011", "BIRTHS2012", "BIRTHS2013", 
                   "BIRTHS2014", "BIRTHS2015", "POPESTIMATE2010", "POPESTIMATE2011",
                  "POPESTIMATE2012", "POPESTIMATE2013", "POPESTIMATE2014", "POPESTIMATE2015"]

df = df[columns_to_keep]
df.head()
```

```python
# Dual Index
df = df.set_index(["STNAME", "CTYNAME"])
df.head()
```

```python
# .loc can take multiple arguments
df.loc["Michigan", "Washtenaw County"]
```
